```python
import random
import math
from collections import defaultdict

class Board(defaultdict):
    """A board for the TicTacToe game."""

    empty = '.'
    off = '#'

    def __init__(self, width=8, height=8, to_move=None, **kwds):
        # Initialize with default value as self.empty
        super().__init__(lambda: self.empty)
        self.width = width
        self.height = height
        self.to_move = to_move
        self.update(**kwds)

    def new(self, changes: dict, **kwds) -> 'Board':
        """Return a new Board with the given changes."""
        board = Board(width=self.width, height=self.height, **kwds)
        board.update(self)
        board.update(changes)
        return board

    def missing(self, loc):
        x, y = loc
        if 0 <= x < self.width and 0 <= y < self.height:
            return self.empty
        else:
            return self.off

    def __hash__(self):
        return hash(tuple(sorted(self.items()))) + hash(self.to_move)

    def __repr__(self):
        def row(y): return ' '.join(self[x, y] for x in range(self.width))
        return '\n'.join(map(row, range(self.height))) + '\n'


class TicTacToe:
    """Play TicTacToe on a board."""

    def __init__(self, height=3, width=3, k=3):
        self.k = k  # k in a row to win
        self.squares = {(x, y) for x in range(width) for y in range(height)}
        self.initial = Board(width=width, height=height, to_move='X')

    def actions(self, board):
        """Legal moves are any square not yet taken."""
        return self.squares - set(board)

    def result(self, board, square):
        """Place a marker for current player on square."""
        player = board.to_move
        board = board.new({square: player}, to_move=('O' if player == 'X' else 'X'))
        return board

    def is_terminal(self, board):
        """A board is a terminal state if it is won or there are no empty squares."""
        if self.check_win(board, 'X') or self.check_win(board, 'O'):
            return True
        if not any(board.missing(square) for square in self.squares):
            return True
        return False

    def check_win(self, board, player):
        """Check if the given player has won the game."""
        for x in range(board.width):
            for y in range(board.height):
                if self.k_in_row(board, player, (x, y), self.k):
                    return True
        return False

    def k_in_row(self, board, player, square, k):
        """True if player has k pieces in a line through square."""
        def in_row(x, y, dx, dy):
            count = 0
            while 0 <= x < board.width and 0 <= y < board.height and board[x, y] == player:
                count += 1
                x += dx
                y += dy
            return count

        x, y = square
        return any(
            in_row(x, y, dx, dy) + in_row(x, y, -dx, -dy) - 1 >= k
            for (dx, dy) in ((0, 1), (1, 0), (1, 1), (1, -1))
        )

    def utility(self, board, player):
        """Return the utility value of the board for the given player."""
        if self.check_win(board, player):
            return 1
        elif self.check_win(board, 'X' if player == 'O' else 'O'):
            return -1
        else:
            return 0

    def display(self, board):
        print(board)


def minimax_search(game, state):
    """Search game tree to determine best move; return (value, move) pair."""
    def max_value(state):
        if game.is_terminal(state):
            return game.utility(state, 'X'), None

        v, move = -math.inf, None
        for a in game.actions(state):
            v2, _ = min_value(game.result(state, a))
            if v2 > v:
                v, move = v2, a

        return v, move

    def min_value(state):
        if game.is_terminal(state):
            return game.utility(state, 'O'), None

        v, move = math.inf, None
        for a in game.actions(state):
            v2, _ = max_value(game.result(state, a))
            if v2 < v:
                v, move = v2, a

        return v, move

    return max_value(state)


def alphabeta_search(game, state):
    """Search game to determine best action; use alpha-beta pruning."""
    def max_value(state, alpha, beta):
        if game.is_terminal(state):
            return game.utility(state, 'X'), None

        v, move = -math.inf, None
        for a in game.actions(state):
            v2, _ = min_value(game.result(state, a), alpha, beta)
            if v2 > v:
                v, move = v2, a
            alpha = max(alpha, v)
            if v >= beta:
                return v, move

        return v, move

    def min_value(state, alpha, beta):
        if game.is_terminal(state):
            return game.utility(state, 'O'), None

        v, move = math.inf, None
        for a in game.actions(state):
            v2, _ = max_value(game.result(state, a), alpha, beta)
            if v2 < v:
                v, move = v2, a
            beta = min(beta, v)
            if v <= alpha:
                return v, move

        return v, move

    return max_value(state, -math.inf, math.inf)


def random_player(game, state):
    """Returns a random action."""
    actions = list(game.actions(state))
    if not actions:
        raise ValueError("No available actions.")
    return random.choice(actions)


def player(search_algorithm):
    """A game player who uses the specified search algorithm."""
    return lambda game, state: search_algorithm(game, state)


def main():
    game = TicTacToe(height=3, width=3, k=3)
    state = game.initial

    print("Initial board:")
    game.display(state)

    # Use MiniMax to determine the best move
    print("MiniMax:")
    value, move = minimax_search(game, state)
    print(f"Best move: {move}, Value: {value}")

    # Apply the move and print the updated board
    state = game.result(state, move)
    game.display(state)

    # Use Alpha-Beta to determine the best move
    print("Alpha-Beta:")
    value, move = alphabeta_search(game, state)
    print(f"Best move: {move}, Value: {value}")

    # Apply the move and print the updated board
    state = game.result(state, move)
    game.display(state)

    # Check if the game is already over
    if not game.is_terminal(state):
        # Use Random player to determine a move
        print("Random Player:")
        try:
            move = random_player(game, state)
            print(f"Random move: {move}")

            # Apply the move and print the updated board
            state = game.result(state, move)
            game.display(state)
        except ValueError as e:
            print(e)
    else:
        print("Game is already over!")


if __name__ == "__main__":
    main()
```

```python
import random
import math
from collections import defaultdict

class Board(defaultdict):
    """A board for the TicTacToe game."""

    empty = '.'
    off = '#'

    def __init__(self, width=8, height=8, to_move=None, **kwds):
        super().__init__(lambda: self.empty)
        self.width = width
        self.height = height
        self.to_move = to_move
        self.update(**kwds)

    def new(self, changes: dict, **kwds) -> 'Board':
        """Return a new Board with the given changes."""
        board = Board(width=self.width, height=self.height, **kwds)
        board.update(self)
        board.update(changes)
        return board

    def missing(self, loc):
        x, y = loc
        if 0 <= x < self.width and 0 <= y < self.height:
            return self.empty
        else:
            return self.off

    def __hash__(self):
        return hash(tuple(sorted(self.items()))) + hash(self.to_move)

    def __repr__(self):
        def row(y): return ' '.join(self[x, y] for x in range(self.width))
        return '\n'.join(map(row, range(self.height))) + '\n'

class TicTacToe:
    """Play TicTacToe on a board."""

    def __init__(self, height=3, width=3, k=3):
        self.k = k  # k in a row to win
        self.squares = {(x, y) for x in range(width) for y in range(height)}
        self.initial = Board(width=width, height=height, to_move='X')

    def actions(self, board):
        """Legal moves are any square not yet taken."""
        return self.squares - set(board)

    def result(self, board, square):
        """Place a marker for current player on square."""
        player = board.to_move
        board = board.new({square: player}, to_move=('O' if player == 'X' else 'X'))
        return board

    def is_terminal(self, board):
        """A board is a terminal state if it is won or there are no empty squares."""
        if self.check_win(board, 'X') or self.check_win(board, 'O'):
            return True
        if not any(board.missing(square) == board.empty for square in self.squares):
            return True
        return False

    def check_win(self, board, player):
        """Check if the given player has won the game."""
        for x in range(board.width):
            for y in range(board.height):
                if self.k_in_row(board, player, (x, y), self.k):
                    return True
        return False

    def k_in_row(self, board, player, square, k):
        """True if player has k pieces in a line through square."""
        def in_row(x, y, dx, dy):
            count = 0
            while 0 <= x < board.width and 0 <= y < board.height and board[x, y] == player:
                count += 1
                x += dx
                y += dy
            return count

        x, y = square
        return any(
            in_row(x, y, dx, dy) + in_row(x, y, -dx, -dy) - 1 >= k
            for (dx, dy) in ((0, 1), (1, 0), (1, 1), (1, -1))
        )

    def utility(self, board, player):
        """Return the utility value of the board for the given player."""
        if self.check_win(board, player):
            return 1
        elif self.check_win(board, 'X' if player == 'O' else 'O'):
            return -1
        else:
            return 0

    def display(self, board):
        print(board)




def minimax_search(game, state):
    """Search game tree to determine best move; return (value, move) pair."""
    def max_value(state):
        if game.is_terminal(state):
            return game.utility(state, 'X'), None

        v, move = -math.inf, None
        for a in game.actions(state):
            v2, _ = min_value(game.result(state, a))
            if v2 > v:
                v, move = v2, a

        return v, move

    def min_value(state):
        if game.is_terminal(state):
            return game.utility(state, 'O'), None

        v, move = math.inf, None
        for a in game.actions(state):
            v2, _ = max_value(game.result(state, a))
            if v2 < v:
                v, move = v2, a

        return v, move

    return max_value(state)


def alphabeta_search(game, state):
    """Search game to determine best action; use alpha-beta pruning."""
    def max_value(state, alpha, beta):
        if game.is_terminal(state):
            return game.utility(state, 'X'), None

        v, move = -math.inf, None
        for a in game.actions(state):
            v2, _ = min_value(game.result(state, a), alpha, beta)
            if v2 > v:
                v, move = v2, a
            alpha = max(alpha, v)
            if v >= beta:
                return v, move

        return v, move

    def min_value(state, alpha, beta):
        if game.is_terminal(state):
            return game.utility(state, 'O'), None

        v, move = math.inf, None
        for a in game.actions(state):
            v2, _ = max_value(game.result(state, a), alpha, beta)
            if v2 < v:
                v, move = v2, a
            beta = min(beta, v)
            if v <= alpha:
                return v, move

        return v, move

    return max_value(state, -math.inf, math.inf)


def random_player(game, state):
    """Returns a random action."""
    actions = list(game.actions(state))
    if not actions:
        raise ValueError("No available actions.")
    return random.choice(actions)


def player(search_algorithm):
    """A game player who uses the specified search algorithm."""
    return lambda game, state: search_algorithm(game, state)[1]  # Return only the move


def play_game(game, players, verbose=False):
    """Play a game between two players using the specified strategies."""
    state = game.initial
    while not game.is_terminal(state):
        if verbose:
            print(f"{state.to_move}'s turn")
            game.display(state)

        actions = list(game.actions(state))
        print(f"Available actions for {state.to_move}: {actions}")  # Debug print

        if not actions:
            print(f"No available actions for player {state.to_move}.")
            break  # Exit the loop if no actions are available

        move = players[state.to_move](game, state)
        print(f"Player {state.to_move} chose move: {move}")  # Debug print

        if move not in actions:
            raise ValueError(f"Player {state.to_move} chose an invalid move {move}.")

        state = game.result(state, move)

    if verbose:
        print("Game Over!")
        game.display(state)

    return game.utility(state, 'X')  # Utility for player 'X'



def main():
    # Strategy 1: X = Random Player, O = Alpha-Beta Search
    print("Strategy 1: X = Random Player, O = Alpha-Beta Search")
    utility = play_game(TicTacToe(), {'X': random_player, 'O': player(alphabeta_search)}, verbose=True)
    print(f"Utility of the final board for X: {utility}")
if __name__ == "__main__":
    main()
```
