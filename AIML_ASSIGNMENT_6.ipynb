```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
np.random.seed(0)
data = {
    'Feature1': np.random.rand(100) * 100,
    'Feature2': np.random.rand(100) * 50,
    'Feature3': np.random.rand(100) * 10,
    'target': np.random.choice(['Class1', 'Class2'], size=100)
}
df = pd.DataFrame(data)
print("First few rows of the dataset:")
print(df.head())
print("\nData types of each feature:")
print(df.dtypes)
print("\nSummary statistics:")
summary = df.describe(include='all', percentiles=[.25, .5, .75, .9])
print(summary)
print("\nColumn names:")
column_names = df.columns.tolist()
print(column_names)
X = df.drop(columns=['target'])
y = df['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.hist(bins=20, figsize=(10, 7))
plt.suptitle('Histograms of Scaled Features')
plt.show()
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(df['Feature1'], df['Feature2'], c=df['target'].apply(lambda x: 0 if x == 'Class1' else 1), cmap='viridis', alpha=0.7)
plt.title('Feature1 vs Feature2')
plt.xlabel('Feature1')
plt.ylabel('Feature2')

plt.subplot(1, 3, 2)
plt.scatter(df['Feature1'], df['Feature3'], c=df['target'].apply(lambda x: 0 if x == 'Class1' else 1), cmap='viridis', alpha=0.7)
plt.title('Feature1 vs Feature3')
plt.xlabel('Feature1')
plt.ylabel('Feature3')

plt.subplot(1, 3, 3)
plt.scatter(df['Feature2'], df['Feature3'], c=df['target'].apply(lambda x: 0 if x == 'Class1' else 1), cmap='viridis', alpha=0.7)
plt.title('Feature2 vs Feature3')
plt.xlabel('Feature2')
plt.ylabel('Feature3')
plt.tight_layout()
plt.suptitle('Scatter Plots of Original Features')
plt.show()
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=0)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print("\nAccuracy of the KNN model:")
print(accuracy)
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

Part 3â€“Evaluate the Classification Quality.

1. Generate the confusion matrix to estimate the correction among features

2. Generate the classification report using classification report


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
np.random.seed(0)
data = {
    'Feature1': np.random.rand(100) * 100,
    'Feature2': np.random.rand(100) * 50,
    'Feature3': np.random.rand(100) * 10,
    'target': np.random.choice(['Class1', 'Class2'], size=100)
}
df = pd.DataFrame(data)
X = df.drop(columns=['target'])
y = df['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.3, random_state=0)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print("\nAccuracy of the KNN model:")
print(accuracy)
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Class1', 'Class2'],
            yticklabels=['Class1', 'Class2'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
report_dict = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report_dict).transpose()
plt.figure(figsize=(10, 6))
sns.heatmap(report_df, annot=True, cmap='coolwarm', fmt='.2f',
            linewidths=.5, linecolor='black')
plt.title('Classification Report')
plt.show()
```
